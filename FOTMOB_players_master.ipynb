```python
# Create an empty list to store all player information
all_players = []

# Iterate through each team ID
for team_id in team_ids:
    try:
        # Make the API request for each team
        url = f'https://www.fotmob.com/api/teams?id={team_id}&ccode3=PRT'
        response = requests.get(url, headers={'User-Agent': 'Chrome/115.0.0.0'})
        
        # Check if the response code is 200 (OK)
        if response.status_code == 200:
            team_info = response.json()

            # Extract the team name and league ID
            team_name = team_info['details']['name']
            league_id = team_info['overview']['table'][0]['data']['leagueId']

            # Extract player information for each role, excluding "coach"
            for role, player_list in team_info['squad']:
                if role == 'coach':
                    continue  # Skip coach information
                for player in player_list:
                    player_id = player['id']
                    player_name = player['name']
                    position = player.get('role', '')  # 'role' might not exist for some players
                    nation = player['ccode']
                    all_players.append([team_id, team_name, league_id, player_id, player_name, position, nation, role])
        else:
            print(f"Error for team ID {team_id}: Response code {response.status_code}")
    except Exception as e:
        print(f"Error for team ID {team_id}: {str(e)}")

# Create a Dataframe from all player data
df = pd.DataFrame(all_players, columns=['Team_ID', 'Team_Name', 'League_ID', 'Player_ID', 'Name', 'Position', 'Nation', 'Role'])

print(df)
```

```python
base_url = 'https://www.fotmob.com/api/newPlayerData?id='

# Iterate through rows of the DataFrame
for index, row in df.iterrows():
    player_id = row['Player_ID']

    try:
        response = requests.get(f'{base_url}{player_id}', headers={'User-Agent': 'Chrome/115.0.0.0'})

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the JSON response
            player_info = response.json()

            # Extract the age information
            age_info = next((info['value']['fallback'] for info in player_info['playerInformation'] if info['title'] == 'Age'), None)
            
            # Add the age information to the 'Age' column in the Dataframe
            df.at[index, 'Age'] = age_info
            
            height_info = next((info['value']['fallback'] for info in player_info['playerInformation'] if info['title'] == 'Height'), None)
            
            # Add the age information to the 'Height' column in the Dataframe
            df.at[index, 'Height'] = height_info
            
            # Extract primary position
            main_position = player_info['positionDescription']['primaryPosition']['label']
            
            # Add the main position to the 'Main_Position' column in the DataFrame
            df.at[index, 'Main Position'] = main_position
            
            # Extract secondary positions
            secondary_positions = [info['strPos']['label'] for info in player_info['positionDescription']['positions'] if not info['isMainPosition']]
            
            # Add secondary positions to separate columns in the Dataframe
            for i, secondary_position in enumerate(secondary_positions, start=1):
                df.at[index, f'Secondary_Position_{i}'] = secondary_position
        else:
            print(f"Failed to fetch data for player ID {player_id}. Status code: {response.status_code}")
    except Exception as e:
        print(f"Error fetching data for player ID {player_id}: {e}")

print(df)
```
